```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page=requests.get("https://presidentofindia.nic.in/former-presidents.htm")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
president_elements=soup.find_all('div',class_="presidentListing")
president_elements
```

```python
president_names = []
president_terms = []
for element in president_elements:
   
    name = element.find('h3').text.strip()
    term = element.find('span', class_='terms').next_sibling.strip()
    
   
    president_names.append(name)
    president_terms.append(term)


data = ({'President Name': president_names,'Term of Office': president_terms})


president_df = pd.DataFrame(data)

president_df
```
