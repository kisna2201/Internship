5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
driver=webdriver.Chrome(r"chromedriver.exe")  
driver.maximize_window()
driver.get("https://www.google.com/maps/")
user_input=input("Enter city name for which coordinates is required : ")

time.sleep(5)
search=driver.find_element(By.XPATH,'//input[@autofocus="autofocus"]')
search.send_keys(user_input)
search_button=driver.find_element(By.ID,'searchbox-searchbutton')
search_button.click()
time.sleep(5)
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")  
driver.maximize_window()
driver.get("https://www.google.com/maps/")
user_input=input("Enter city name for which coordinates is required : ")

time.sleep(5)
search=driver.find_element(By.XPATH,'//input[@autofocus="autofocus"]')
search.click()
search.send_keys(user_input)
search_button=driver.find_element(By.ID,'searchbox-searchbutton')
search_button.click()
time.sleep(5)
lat=[]
long=[]
try:
    url_string = driver.current_url
    print("URL Extracted :",url_string)
    lat_long = re.findall(r'@(.*)data',url_string)
    if len(lat_long):
        lat_long_list = lat_long[0].split(",")
        if len(lat_long_list)>=2:
            lat.append(lat_long_list[0])
            long.append(lat_long_list[1])
except Exception as e:
    print("error:",str(e))

time.sleep(2)
```

```python
location_data=pd.DataFrame({'Enter Location':user_input,'latitude':lat,'longitude':long})
location_data
```

```python
rank=[]
name=[]
net_worth=[]
age=[]
citizenship=[]
source=[]
industry=[]

time.sleep(3)
start=0
end=13

for page in range(start,end):
    ranks=driver.find_elements(By.XPATH,'//div[@class="Table_rank___YBhk Table_dataCell__2QCve"]')
    try:
        for i in ranks[0:10]:
            b_ranks=i.text
            rank.append(b_ranks)
    except NoSuchElementException:
        rank.append("-")
    
    names=driver.find_elements(By.XPATH,'//div[@class="TableRow_row__L-0Km TableRow_activeRow__g4oSF"]/div[2]')
    try:
        for i in names[0:10]:
            b_names=i.text
            name.append(b_names)
    except NoSuchElementException:
        name.append("-")
        
    netWorth=driver.find_elements(By.XPATH,'//div[@class="TableRow_row__L-0Km TableRow_activeRow__g4oSF"]/div[3]')
    try:
        for i in netWorth[0:10]:
            b_networth=i.text
            net_worth.append(b_networth)
    except NoSuchElementException:
        net_worth.append("-")
    
    ages=driver.find_elements(By.XPATH,'/div[@class="TableRow_row__L-0Km TableRow_activeRow__g4oSF"]//div[4]')
    try:
        for i in ages[0:10]:
            b_age=i.text
            age.append(b_age)
    except NoSuchElementException:
        age.append("-")
    
    citizenships=driver.find_elements(By.XPATH,'//div[@class="TableRow_row__L-0Km TableRow_activeRow__g4oSF"]//div[5]')
    try:
        for i in citizenships[0:10]:
            b_citizenship=i.text
            citizenship.append(b_citizenship)
    except NoSuchElementException:
        citizenship.append("-")
    
    sources=driver.find_elements(By.XPATH,'//div[@class="TableRow_row__L-0Km TableRow_activeRow__g4oSF"]//div[6]')
    try:
        for i in sources[0:10]:
            b_source=i.text
            source.append(b_source)
    except NoSuchElementException:
        source.append("-")
        
    industries=driver.find_elements(By.XPATH,'//div[@class="TableRow_row__L-0Km TableRow_activeRow__g4oSF"]//div[7]')
    try:
        for i in industries[0:10]:
            b_industries=i.text
            industry.append(b_industries)
    except NoSuchElementException:
          industry.append("-")
```

Q 9 Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.


```python
driver = webdriver.Chrome(r"chromedriver.exe")
driver.maximize_window()
```

```python
try:

    driver.get("https://www.hostelworld.com/")

    loc = driver.find_element(
        By.XPATH,
        "/html/body/div[3]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/div/input",
    )

    loc.send_keys("london")
    search_btn = driver.find_element(
        By.XPATH,
        "/html/body/div[3]/div/div/div[2]/div[1]/div/div/div[4]/div/div[2]/div/div[5]/button",
    )
    search_btn.click()

except Exception as e:

    driver.get(
        "https://www.hostelworld.com/s?q=London, England&country=England&city=London&type=city&id=3&from=2022-07-28&to=2022-07-31&guests=2&HostelNumber"
    )


time.sleep(3)
```

```python
hostel_name = []
location = []
ratings = []
total_reviews = []
overall_reviews = []
private_price = []
dorm_price = []

hostel = driver.find_elements(By.XPATH, "//h2[@class='title title-6']")
try:
    for i in hostel:
        hostel_name.append(i.text)
except:
    hostel_name.append("-")


distance = driver.find_elements(By.XPATH, "//span[@class='description']")
try:
    for i in distance:
        location.append(i.text)
except:
    location.append("-")


information = driver.find_elements(By.XPATH, "//div[@class='info']/div[2]")
try:
    for i in information:
        ratings.append(i.text.split("\n")[0])
except:
    ratings.append("-")

reviews = driver.find_elements(By.XPATH, "//div[@class='reviews']")
try:
    for i in reviews:
        total_reviews.append(i.text)
except:
    total_reviews.append("-")


review = driver.find_elements(By.XPATH, "//div[@class='keyword']/span")
try:
    for i in review:
        overall_reviews.append(i.text)
except:
    overall_reviews.append("-")


private = driver.find_elements(By.XPATH, "//a[@class='prices']/div[1]")
try:
    for i in private:
        private_price.append(i.text.replace("\n", ""))
except:
    private_price.append("-")


dorm = driver.find_elements(By.XPATH, "//a[@class='prices']/div[2]")
for i in dorm:
    dorm_price.append(i.text.replace("\n", ""))


property_description = []
prop_url = []
url = driver.find_elements(By.XPATH, "//h2[@class='title title-6']/a")
for i in url:
    prop_url.append(i.get_attribute("href"))

for i in prop_url:
    driver.get(i)
    time.sleep(5)
    try:
        description = driver.find_element(
            By.XPATH,
            "/html/body/div[3]/div/div/div[2]/section/div[6]/div/div[2]/div/div/div[2]",
        )
        property_description.append(description.text)
    except NoSuchElementException:
        property_description.append("-")

time.sleep(5)
driver.close()
```

```python
print(len(hostel_name),len(location),len(ratings),len(total_reviews),len(overall_reviews),len(private_price),len(dorm_price),len(property_description),len(prop_url))

```

```python
available_hostel=pd.DataFrame({"Hostel_Name":hostel_name,"Loction":location,"Rating":ratings,"Total Review":total_reviews,"Overall_Review":overall_reviews[0:25],"Private Price":private_price[0:25],"Drom_Price":dorm_price,"Description":property_description,"URL":prop_url})
available_hostel 
```
